```python
import pandas as pd
train_df = pd.read_csv("poem_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("poem_test.csv", encoding="utf-8")
test_df
```

```python
import jieba
s = "天門中斷楚江開，碧水東流至此回。兩岸青山相對出，孤帆一片日邊來。"
s = s.replace("\r","").replace("\n","")
" ".join(jieba.cut(s))
```

```python
# apply: 對整行坐同一操作
def poemcut(s):
    s = s.replace("\r","").replace("\n","")
    return " ".join(jieba.cut(s))

# 給步驟(poemcut)
X_train = train_df["內容"].apply(poemcut)
X_test = test_df["內容"].apply(poemcut)
```

```python
X_train
```

```python
X_test
```

```python
# ML & DL 的函式庫都要求數字
train_df["作者"].value_counts()
```

```python
u = train_df["作者"].unique()
u
```

```python
trans = {v:i for i, v in enumerate(u)}
trans
```

```python
reverse = {i:v for i, v in enumerate(u)}
reverse
```

```python
y_train = train_df["作者"].map(trans)
y_train
```

```python
y_test = test_df["作者"].map(trans)
y_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X_train_vec = vec.fit_transform(X_train)
X_test_vec = vec.transform(X_test)
```

```python
X_train_vec
```

```python
X_test_vec
```

```python
word2idx = vec.vocabulary_
word2idx
```

```python
idx2word = {v:k for k, v in word2idx.items()}
idx2word
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(X_test_vec)
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pre)
```

```python
i = [(name + "(正確)") for name in u]
c = [(name + "(預測)") for name in u]
pd.DataFrame(confusion_matrix(y_test, pre), columns=c, index=i)
```

```python
p = input("寫一首詩: ")
p = poemcut(p)
p = vec.transform([p])
p
```

```python
proba = clf.predict_proba(p)
proba
```
