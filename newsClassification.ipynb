```python
import glob
import os
import pandas as pd
train_Dir = "chinese_news_trans"
test_Dir = "chinese_news_test"
def read_data(dirname):
    filenames = glob.glob(dirname+"/*/*.txt")
    contents = []
    answers = []
    for filename in filenames:
        with open(filename, "r", encoding="utf-8") as f:
            news = f.read()
        contents.append(news)
        dn = os.path.split(filename)[0]
        dn = os.path.split(dn)[-1]
        answers.append(dn)
    df = pd.DataFrame({"content": contents, "ans": answers})
    return df
# answers
```

```python
train_df = read_data(train_Dir)
test_df = read_data(test_Dir)
```

```python
import jieba
jieba.set_dictionary("big_dict.txt")
```

```python
def newscut(s):
    s = s.replace("\r","").replace("\n","")
    return " ".join(jieba.cut(s))

# 給步驟(poemcut)
X_train = train_df["content"].apply(newscut)
X_test = test_df["content"].apply(newscut)
```

```python
u = train_df["ans"].unique()
u
```

```python
label2idx = {v:i for i, v in enumerate(u)}
label2idx
```

```python
idx2label = {i:v for i, v in enumerate(u)}
idx2label
```

```python
y_train = train_df["ans"].map(label2idx)
y_train
```

```python
y_test = test_df["ans"].map(label2idx)
y_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X_train_vec = vec.fit_transform(X_train)
X_test_vec = vec.transform(X_test)
```

```python
X_train_vec
```

```python
X_test_vec
```

```python
word2idx = vec.vocabulary_
word2idx
```

```python
idx2word = {v:k for k, v in word2idx.items()}
idx2word
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(X_test_vec)
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pre)
```

```python
i = [(name + "(正確)") for name in u]
c = [(name + "(預測)") for name in u]
pd.DataFrame(confusion_matrix(y_test, pre), columns=c, index=i)
```
