# Missing value

```python
# make a list of the variables that contain missing values
var_with_na = [var for var in data.columns if  data[var].isnull().sum()>1]

# print the variable name and the percentage of missing values
for var in var_with_na:
    print(var, np.round(data[var].isnull().mean(), 3), '% missing values')
```

#### Relationship between values  being missing and SalePrice

```python
from numpy import median
def analyse_na_value(df, var) :
    df = df.copy()
    
    #let's make a variable that indicates 1 if the observation was missing or 0 otherwise
    df[var] = np.where(df[var].isnull(),1,0)
    
    #let's calculate the mean SalePrice where the information is missing or present
    #df.groupby(var)['SalePrice'].median().plot.bar()
    
    sns.barplot(x=var,y='SalePrice', data=df, estimator=median)
    plt.title(var)
    plt.show()

for var in var_with_na:
    analyse_na_value(data,var)
    
```

#### There has been a drop in the value of the house, that is unusual

```python
# let's explore the relationship between the year variables and the house price in a bit of more details
def analyse_year_vars(df, var):
    df = df.copy()
    df[var] = df['YrSold']-df[var]
    
    #plt.scatter(df[var], df['SalePrice'])
    sns.scatterplot(x=var, y='SalePrice', data=df)
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in year_vars:
    if var != 'YrSold':
        analyse_year_vars(data,var)
```

#### Discrete variables

```python
# list of discrete variables
discrete_vars = [var for var in num_vars if len(data[var].unique())<20 and var not in year_vars+["Id"]]

print('Number of discrete variables : ',len(discrete_vars))
```

```python
data[discrete_vars].head()
```

```python
def analyse_discrete(df, var):
    df = df.copy()
    #df.groupby(var)['SalePrice'].median().plot.bar()
    sns.barplot(x=var, y='SalePrice', data=df, estimator=median)
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data, var)
```

#### Continuous variables

```python
cont_vars = [var for var in num_vars if var not in discrete_vars+year_vars+['Id']]

print('Number of continuous variables : ',len(cont_vars))
```

```python
data[cont_vars].head()
```

```python
def analyse_continuous(df,var):
    df = df.copy()
    
    #df[var].hist(bins=20)
    sns.distplot(df[var].dropna(), kde=False, bins=20)
    plt.ylabel('Number of house')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_continuous(data,var)
```

```python
# for linear regression , we transform the distribution to normal

def analyse_transformed_continuous(df,var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else: 
        df[var] = np.log(df[var])
        #df[var].hist(bins=20)
        sns.distplot(df[var].dropna(), bins=20, kde=False ,color='y')
        plt.ylabel('Numbers of housees')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
        
for var in cont_vars:
    analyse_transformed_continuous(data, var)
```

```python
# let's explore the relationship between the house price and the transformed varibles

def transform_analyse_continuous(df,var):
    df=df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df['SalePrice'] = np.log(df['SalePrice'])
        #plt.scatter(df[var],df['SalePrice'])
        sns.scatterplot(var, 'SalePrice', data=df)
        plt.ylabel('SalePrice')
        plt.xlabel(var)
        plt.show()
        
for var in cont_vars:
    transform_analyse_continuous(data,var)
```

# Outliers

```python
 def finf_outliers(df, vars):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        #df.boxplot(column=var)
        sns.boxplot(var, data=df, orient='v')
        plt.title(var)
        plt.ylabel(var)
        plt.show()
        
for var in cont_vars:
    finf_outliers(data,var)
```

### Rare labels 

```python
def analyse_rare_labels(df, var, rare_perc):
    df=df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp < rare_perc]

for var in cat_vars:
    print(analyse_rare_labels(data,var,0.01))
    print()
```
