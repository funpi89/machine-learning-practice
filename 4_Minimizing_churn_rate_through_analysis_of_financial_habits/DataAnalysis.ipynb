# Removing NaN

```python
dataset.isna().any()
```

```python
dataset.isna().sum()
```

```python
# age only have 4 NaN , 直接除掉這4
dataset = dataset[pd.notnull(dataset['age'])]
```

```python
# credit_score & rewards_earned NaN 比例太多, noise too big, 直接去除
dataset = dataset.drop(columns=['credit_score','rewards_earned'])
```

# Histograms

```python
dataset2 = dataset.drop(columns=['user','churn','housing','zodiac_sign','payment_type'])
# user , churn 不是feature , housing,zodiac_sign,payment_type 不是數值,先去掉(或是one hot encoding)才可以做直方圖
dataset2.shape
```

```python
plt.suptitle('Histograms of Numerical columns',fontsize=20)
plt.figure(figsize=(15,30))
for i in range(1,dataset2.shape[1]+1):
    plt.subplot(8,3,i)
    f=plt.gca()
    f.set_title(dataset2.columns.values[i-1])
    
    vals = np.size(dataset2.iloc[:,i-1].unique())
    #sn.countplot(dataset2.iloc[:,i-1],color="#3F5D7D")
    plt.hist(dataset2.iloc[:,i-1],bins=vals,color='#3F5D7D')
#plt.tight_layout(rect=[0,0.03,1,0.95])
```

# Pie Plots

```python
# 使用只有2位元(0,1)的選項
dataset2 = dataset[['housing',
       'app_downloaded', 'web_user', 'app_web_user', 'ios_user',
       'android_user', 'registered_phones', 'payment_type', 'waiting_4_loan',
       'cancelled_loan', 'received_loan', 'rejected_loan', 'zodiac_sign',
       'left_for_two_month_plus', 'left_for_one_month',
        'is_referred']]

plt.suptitle('Pie Chart Distributions',fontsize=20)
plt.figure(figsize=(15,30))
for i in range(1,dataset2.shape[1]+1):
    plt.subplot(8,3,i)
    f=plt.gca()
    f.set_title(dataset2.columns.values[i-1])
    values = dataset2.iloc[:,i-1].value_counts(normalize = True).values
    index = dataset2.iloc[:,i-1].value_counts(normalize = True).index
    
    plt.pie(values, labels = index, autopct='%1.1f%%')
    plt.axis('equal')
```

# Correlation Plot

```python
# 移除category者
dataset.drop(columns=['churn','user','housing','payment_type',
            'zodiac_sign']).corrwith(dataset.churn).plot.bar(
figsize = (20,10), title = 'Correlation with Response Variable',
fontsize = 15, rot = 45, grid = True)
```

```python
# age : 負相關,越年輕就越可能流失
# deposit, widthdrawal, purchase : 活動越多越不會流失
# cc_taken : 可能信用卡方案或用戶體驗不好導致信用卡與流失率呈正相關
# cc_recommaended : 給客戶越多建議,流失率會越低
#
```

# Correlation Matrix

```python
sn.set(style="white")

corr = dataset.drop(columns=['user','churn']).corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f ,ax = plt.subplots(figsize=(18,15))

cmap = sn.diverging_palette(220, 10, as_cmap=True)

sn.heatmap(corr,mask=mask, cmap=cmap, vmax=.3, center=0,
          square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
