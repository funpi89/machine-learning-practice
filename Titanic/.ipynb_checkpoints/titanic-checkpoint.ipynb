確認整個data的遺失值

```python
data.info()
```

```python
data['Embarked'] = data['Embarked'].fillna('S')  #遺失值只有一個,填最多者
data['Fare'] = data['Fare'].fillna(data['Fare'].mean()) #遺失值只有一個,填平均票價
```

```python
data.info()
```

```python
data['Cabin'].head()
```

```python
data["Cabin"] = data['Cabin'].apply(lambda x : str(x)[0] if not pd.isnull(x) else 'NoCabin') #遺失值瑱N
data['Cabin'].unique()
```

```python
data.info()
```

```python
sns.countplot(data['Cabin'],hue=data['Survived'])
```

將非數值的特徵轉為數值

```python
data['Sex'] = data['Sex'].astype('category').cat.codes
data['Embarked'] = data['Embarked'].astype('category').cat.codes
data['Pclass'] = data['Pclass'].astype('category').cat.codes
data['Title2'] = data['Title2'].astype('category').cat.codes
data['Cabin'] = data['Cabin'].astype('category').cat.codes
data['Ticket'] = data['Ticket'].astype('category').cat.codes
```

先預測年齡再將預測出的年齡填入年齡的遺失值

```python
dataAgeNull = data[data["Age"].isnull()]
dataAgeNotNull = data[data["Age"].notnull()]
#先移除偏遠值 (remove_outer : 船票價 或 家庭人數 之與平均值的差距大於其4被標準差者)
remove_outer = dataAgeNotNull[(np.abs(dataAgeNotNull['Fare']-dataAgeNotNull['Fare'].mean())<(4*dataAgeNotNull['Fare'].std())) | (np.abs(dataAgeNotNull['family']-dataAgeNotNull['family'].mean())<(4*dataAgeNotNull['family'].std()))]


rfAge = RandomForestRegressor(n_estimators=2000,random_state=42)
agecol = ['Embarked', 'Fare', 'Pclass', 'Sex', 'family', 'Title2','Cabin','Ticket']
rfAge.fit(dataAgeNotNull[agecol],dataAgeNotNull['Age'])

ages = rfAge.predict(X=dataAgeNull[agecol])
dataAgeNull['Age']=ages

data = dataAgeNull.append(dataAgeNotNull)

```

```python
data = data.sort_values(by=["PassengerId"])
data.reset_index(inplace=True,drop=True)
data
```

```python
dataTrain = data[data['Survived'].notnull()]
Train_col = ['Survived','Age','Embarked', 'Fare', 'Pclass', 'Sex', 'family', 'Title2','Cabin']
dataTrain = dataTrain[Train_col]
dataTrain.head()
```

```python
dataTest = data[data['Survived'].isnull()]
Test_col = ['Age','Embarked', 'Fare', 'Pclass', 'Sex', 'family', 'Title2','Cabin']
dataTest = dataTest[Test_col]
dataTest.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion='gini',n_estimators=1000,min_samples_split=12,min_samples_leaf=1,oob_score=True,random_state=1,n_jobs=-1)
rf.fit(dataTrain.iloc[:,1:],dataTrain.iloc[:,0])
print("%.4f" % rf.oob_score)
```

查看特徵的重要性

```python
pd.concat((pd.DataFrame(dataTrain.iloc[:,1:].columns,columns=['variable']),
           pd.DataFrame(rf.feature_importances_,columns=['import'])),axis=1).sort_values(by='import',ascending=False)[:20]
```

Submit

```python
rf_res = rf.predict(dataTest)
submit['Survived'] = rf_res
submit['Survived'] = submit['Survived'].astype(int)
submit.to_csv('submit4.csv',index=False)
submit.head()
```
